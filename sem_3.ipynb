1.1 Создать новый признак ram_size, который будет
содержать количество оперативной памяти

```python
def kolk(value): 
    return int(value.split()[0])
```

```python
df["ram_size"] = df["ram_gb"].apply(kolk)
```

1.2. Создать новый признак all_size, который будет содержать сумму памяти ssd и hdd

```python
def calc_all_size(row):
    ssd_size = int(row["ssd"].split()[0])
    hdd_size = int(row["hdd"].split()[0])
    return ssd_size+hdd_size
```

```python
df["all_size"]=df.apply(calc_all_size, axis=1)
```
